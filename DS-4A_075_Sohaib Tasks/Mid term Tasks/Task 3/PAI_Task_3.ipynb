```python
def water_jug(jug1_capacity, jug2_capacity, target_amount):

    initial_state = (0, 0)
    target_state = target_amount


    stack = [(initial_state, [])]
    visited = set()

    while stack:
        current_state, path = stack.pop()

        if current_state in visited:
            continue
        visited.add(current_state)

        jug1_current, jug2_current = current_state


        if jug1_current == target_state or jug2_current == target_state:
            for step in path:
                print(step)
            print(f"Target ({target_state}) reached at state: {current_state}\n")
            return True

        if jug1_current < jug1_capacity:
            new_state = (jug1_capacity, jug2_current)
            rule = f"Fill Jug 1 (from {current_state} to {new_state})\n"
            stack.append((new_state, path + [rule]))


        if jug2_current < jug2_capacity:
            new_state = (jug1_current, jug2_capacity)
            rule = f"Fill Jug 2 (from {current_state} to {new_state})\n"
            stack.append((new_state, path + [rule]))


        if jug1_current > 0:
            new_state = (0, jug2_current)
            rule = f"Empty Jug 1 (from {current_state} to {new_state})\n"
            stack.append((new_state, path + [rule]))


        if jug2_current > 0:
            new_state = (jug1_current, 0)
            rule = f"Empty Jug 2 (from {current_state} to {new_state})\n"
            stack.append((new_state, path + [rule]))


        if jug1_current > 0 and jug2_current < jug2_capacity:
            pour_amount = min(jug1_current, jug2_capacity - jug2_current)
            new_state = (jug1_current - pour_amount, jug2_current + pour_amount)
            rule = f"Pour Jug 1 to Jug 2 (from {current_state} to {new_state})\n"
            stack.append((new_state, path + [rule]))


        if jug2_current > 0 and jug1_current < jug1_capacity:
            pour_amount = min(jug2_current, jug1_capacity - jug1_current)
            new_state = (jug1_current + pour_amount, jug2_current - pour_amount)
            rule = f"Pour Jug 2 to Jug 1 (from {current_state} to {new_state})\n"
            stack.append((new_state, path + [rule]))

    print("Target not reachable.")
    return False
```

```python
print("Solving Water Jug Problem for 4-liter and 3-liter jugs, target 2 liters:\n")
water_jug(4, 3, 2)

print("\nSolving Water Jug Problem for 5-liter and 3-liter jugs, target 4 liters:\n")
water_jug(5, 3, 4)
```
